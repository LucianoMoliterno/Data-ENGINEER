# Transformación de datos

En los casos ideales es que los daton lleguen con los datos en la forma correcta para que los podamos utilizar para hacer gráficos. Como en la vida real las cosas no son así y en ocasiones será necesario cambiar ciertas cosas de los conjuntos de datos como cambiar el nombre a las columnas podemos utilizar un paquete llamado `dplyr`.

En este capítulo utilizaremos un conjunto de datos de cuelos que salen de Nueva York en el 2013.

```r
install.packages("nycflights13")
library(nycflights13)
```

```r
flights
```

```r
glimpse(flights)
```

```r
help(flights)
```

```r
# |> tuberia
# Pasa lo que se haya hecho en la izquiera y lo pasa a su función
# De la derecha
flights |>
  filter(dest == "IAH") |>
  group_by(year, month, day) |>
  summarize(
    arr_delay = mean(arr_delay, na.rm = T)
  )
```

## Filas

### filter

```r
flights |>
  filter(dep_delay > 120)
```

```r
flights |>
  filter(month == 1 & day == 1)
```

```r
# %in% la variable sea igual a un valor de la derecha
flights |>
  filter(month %in% c(1, 2))
```

## arrange()

Cambia los valores de las filas basado en los valores de las columnas.

```r
flights |>
  arrange(year, month, day, dep_time)
```

```r
flights |>
  arrange(desc(dep_delay))
```

## distinct()

Encuentra todas las filas únicas en el conjunto de datos

```r
flights |>
  distinct(origin, dest)
```

```r
flights |>
  distinct(origin, dest, .keep_all = T)
```

```r
flights |>
  count(origin, dest, sort = T)
```

### Columnas

#### mutate()
Añadir nuevas columnas que son calculadas en base a los datos de las columnas existintes.

```r
flights |>
  mutate(
    recupero_vuelo = dep_delay - arr_delay,
    velocidad = (distance * 1.609344) / (air_time / 60)
  )
```

```r
# Añadimos las columnas al inicio
flights |>
  mutate(
    recupero_vuelo = dep_delay - arr_delay,
    velocidad = (distance * 1.609344) / (air_time / 60),
    .before = 1
  )
```

```r
flights |>
  mutate(
    recupero_vuelo = dep_delay - arr_delay,
    velocidad = (distance * 1.609344) / (air_time / 60),
    .after = day
  )
```

```r
flights |>
  mutate(
    recupero_vuelo = dep_delay - arr_delay,
    velocidad = (distance * 1.609344) / (air_time / 60),
    .keep = "used"
  )
```

### relocate()

recolocar las columnas importantes al inicio, por default las mueve al inicio

```r
flights |>
  relocate(time_hour, air_time)
```

```r
flights |>
  relocate(year:dep_time, .after = time_hour)
```

```r

flights |>
  relocate(starts_with("arr"), .before = dep_time)
```

## Grupos

### group_by()

Divide el conjunto de datos en grupos para hacer más fácil el análisis

```r
flights |>
  group_by(month)
```

### summarize()

Se utiliza para cálcular un valor estadístico que haga un resumen del conjunto.

```r
flights |>
  group_by(month) |>
  summarize(
    avg_delay = mean(dep_delay, na.rm = T)
  )
```

```r
# la cantidad de filas en cada grupo
flights |>
  group_by(month) |>
  summarize(
    avg_delay = mean(dep_delay, na.rm = T),
    n = n()
  )
```

## slice_Functions

- df |> slice_head(n = 1) La primer fila del grupo
- df |> slice_tail(n = 1) La últime fila del grupo
- df |> slice_min(x, n = 1) La fila del valor más pequeño de la columna
- df |> slice_min(x, n = 1) La fila del valor más grande de la columna
- df |> slice_sample(x, n = 1) Toma una fila aleatoria

```r
# Selecciona los vualos con más retraso en cada destino
flights |>
  group_by(dest) |>
  slice_max(arr_delay, n = 1) |>
  relocate(dest)
```

```r
flights |>
  group_by(dest) |>
  slice_min(arr_delay, n = 1) |>
  relocate(dest)
```

### Agrupar por múltiples variables

```r
daily <- flights |>
  group_by(year, month, day)
```

```r
daily |>
  summarize(n = n())
```

### .by

`dplyr` tiene a partir de la versión 1.1.0 un sintáxis especial para agrupar los datos

```r
flights |>
  summarize(
    delay = mean(dep_delay, na.rm = T),
    n = n(),
    .by = month
  )
```

```r
flights |>
  summarize(
    delay = mean(dep_delay, na.rm = TRUE),
    n = n(),
    .by = c(origin, dest)
  )
```

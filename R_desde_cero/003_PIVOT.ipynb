# Datos ordenanos

Organizar los datos es muy importante y en R existe un sistema al que se le llama `tidy data` (datos ordenados). Esto sirve para tener los datos ordenados y en el formato correcto para poder usarlos. Utilizando estos métodos podremos pasar más tiempo respondiendo preguntas que preocupandonos por los datos en sí.

## Datos alargados

La mayoria de los datos no se encuentran ordenados, esto se debe a dos razones:

- Data normalmente se organiza para conseguir otros objetivos al hacer ciertos análisis.

- La mayoria de las personas no están familiarizados con los principios de datos ordenados

### Datos como nombre en las columnas

```r
billboard
```

Las primeras tres columnas de este conjunto de datos son columnas que describen a las canciones. Las restantes columnas describen el rango de esa canción en cada semana. Cambiaremos las información de esas columnas a una sola.

```r
billboard |>
  pivot_longer(
    cols = starts_with("wk"),
    names_to = "week",
    values_to = "rank"
  )
```

```r
billboard_pivot <- billboard |>
  pivot_longer(
    cols = starts_with("wk"),
    names_to = "week",
    values_to = "rank",
    values_drop_na = T
  )
```

```r
glimpse(billboard_pivot)
```

```r
billboard_pivot <- billboard |>
  pivot_longer(
    cols = starts_with("wk"),
    names_to = "week",
    values_to = "rank",
    values_drop_na = T
  ) |>
    mutate(
      week = parse_number(week)
    )
```

```r
glimpse(billboard_pivot)
```

```r
billboard_pivot |>
  ggplot(aes(x = week, y = rank, group = track)) +
  geom_line(alpha = 0.15) +
  scale_y_reverse()
```

## Muchas variables en los nombres de las columnas

```r
who2
```

Este dataset tiene información sobre los diagnósticos de tuberculosis recolectados por la Organización Mundial de la Salud. Estas columnas tienen 56 columnas que que le siguen `country` y `year`. En estas columnas el nombre esta dividido por `_`. La primera parte del nombre tiene `sp/rel/ep`, describe el métpdo usado para el diagnostico; la segunda parte `m/f` tiene el género y la última parte `014/1524/2534/3544/4554/65` es el rango de edades.

```r
who2 |>
  pivot_longer(
    cols = !(country:year),
    names_to = c("diagnosis", "gender", "age"),
    names_sep = "_",
    values_to = "count"
  )
```

Este conjunto de datos tiene datos sobre 5 familias, con los nombres y fechas de nacimiento de los dos niños. El reto en este conjunto de datos es que los nombres de las columnas tienen dos variables (`dob`, `name`) y los valores

```r
household
```

En este caso utilizaremos un valor especial `.value`, este lo que hace es que sobreescribe el funcionamiento del argumento `values_to` y lo que hace es utilizar el primer componente del nombre de la columna pivotada como nombre de variable en la salida.

```r
household |>
  pivot_longer(
    cols = !family,
    names_to = c(".value", "child"),
    names_sep = "_",
    values_drop_na = T
  )
```

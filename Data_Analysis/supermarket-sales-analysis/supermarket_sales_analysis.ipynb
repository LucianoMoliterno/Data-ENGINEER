```python
#IMPORTO LIBRERIAS (PANDAS,MATPLOT,SEABORN)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1️⃣ Cargo dataset
df = pd.read_csv("supermarket_sales.csv")
```

```python
# 2️⃣ Limpieza de datos
df.drop_duplicates(inplace=True)  # Eliminar duplicados
df.dropna(inplace=True)  # Eliminar valores nulos
```

```python
# Convierte fechas
df["Date"] = pd.to_datetime(df["Date"])
df["Day"] = df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month_name()
```

```python
# 3️⃣ Análisis Exploratorio
print(df.info())
print(df.describe())
```

```python
# Ventas por tipo de producto
plt.figure(figsize=(10,5))
sns.countplot(data=df, y="Product line", order=df["Product line"].value_counts().index, palette="viridis")
plt.title("Ventas por Categoría de Producto")
plt.show()
```

```python
# Tendencia de ventas por mes
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="Date", y="Total", hue="City", palette="tab10")
plt.title("Tendencia de Ventas por Ciudad")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4️⃣ Conclusiones clave
top_product = df["Product line"].value_counts().idxmax()
top_city = df.groupby("City")["Total"].sum().idxmax()
print(f"📌 Producto más vendido: {top_product}")
print(f"📌 Ciudad con más ventas: {top_city}")
```
